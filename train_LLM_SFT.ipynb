```python
from transformers import BitsAndBytesConfig
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "MLP-KTLim/llama-3-Korean-Bllossom-8B"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model_nf4 = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config)
```

```python
PROMPT = '''You are a helpful AI assistant. Please answer the user's questions kindly. 당신은 유능한 AI 어시스턴트 입니다. 사용자의 질문에 대해 친절하게 답변해주세요.'''
instruction = "서울의 유명한 관광 코스를 만들어줄래?"
```

```python
messages = [
    {"role": "system", "content": f"{PROMPT}"},
    {"role": "user", "content": f"{instruction}"}
    ]

input_ids = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt"
).to(model_nf4.device)

terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
```

```python
def formatting_prompts_func(example):
    output_texts = []
    for i in range(len(example)):
        text = f"### Question: {example['question'][i]}\n ### Answer: {example['answer'][i]}"
        output_texts.append(text)
    return output_texts
```

```python
def print_gpu_utilization():
    if torch.cuda.is_available():
        used_memory = torch.cuda.memory_allocated(0) / 1024**3
        print(f"GPU RAM Usage: {used_memory:.3f} GB")
    else :
        print("GPU not available")
```

```python
def clear_gpu_cache():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        print("GPU cache cleared")
    else :
        print("GPU not available")
```

```python
def estimate_memory_of_gradients(model):
    total_memory = 0
    for name, param in model.named_parameters():
        if param.requires_grad:
            param_memory = param.element_size() * param.nelement()
            total_memory += param_memory
    return total_memory
```

```python
def estimate_memory_of_optimizer(optimizer):
    total_memory = 0
    for group in optimizer.param_groups:
        for param in group['params']:
            param_memory = param.element_size() * param.nelement()
            total_memory += param_memory
    return total_memory
```

```python
import gc

def clean_memory():
    if 'model' in globals():
        del globals()['model']
    if 'dataset' in globals():
        del globals()['dataset']
    gc.collect()
    torch.cuda.empty_cache()
```

```python
train_datas = formatting_prompts_func(data)
```

```python
input_ids = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt"
).to(model_nf4.device)

terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
```

```python
from torch.utils.data import DataLoader
from transformers import AdamW
from torch.optim import SGD

BATCH_SIZE = 1

train_dataloader = DataLoader(train_datas, batch_size=BATCH_SIZE)

# optimizer = AdamW(model_nf4.parameters())
optimizer = SGD(model_nf4.parameters())
model_nf4.train()
gpu_utilization_printed = False

for step, batch in enumerate(train_dataloader, start=1):

    # print(len(batch))
    input_ids = tokenizer(batch, return_tensors="pt", padding=True).to(model_nf4.device)['input_ids']
    outputs = model_nf4(input_ids)
    loss = outputs.loss
    loss.backward()
    optimizer.step()
    optimizer_memory = estimate_memory_of_optimizer(optimizer)
    gradient_memory = estimate_memory_of_gradients(model_nf4)
    print('Optimizer memory : ', optimizer_memory)
    print('Gradient momory : ', gradient_memory)
    # print(f"Step {step}, Loss: {loss.item()}"
    optimizer.zero_grad()
    break

```

```python
from trl import SFTConfig, SFTTrainer, DataCollatorForCompletionOnlyLM
from peft import LoraConfig, get_peft_model

response_template = " ### Answer:"
collator = DataCollatorForCompletionOnlyLM(response_template, tokenizer=tokenizer)

lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["self_attn.q_proj", "self_attn.v_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM",
)

model_peft = get_peft_model(model_nf4, lora_config)

trainer = SFTTrainer(
    model_peft,
    train_dataset=data,
    args=SFTConfig(output_dir="/tmp", report_to=None),
    formatting_func=formatting_prompts_func,
    data_collator=collator,
)

trainer.train()
```
